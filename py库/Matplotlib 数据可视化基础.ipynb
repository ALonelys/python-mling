```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)    #  一个范围切成100段
print(x)
```

```python
y=np.sin(x)     #创建余弦函数
print(y)
```

###  plt.plot(x,y)  x是横轴,y是纵轴

```python
plt.plot(x,y)  #第一个为横坐标，第二个人为纵坐标   
plt.show()   #需要一个show 函数，可以输出图形,其实是折线图，但是点多就平滑了
```

```python
cosy=np.cos(x)
print(cosy.shape)
```

```python
siny=y.copy()
print(siny)
print(cosy)
```

### plt.show() 可以输出多条曲线，plt.plot()可以一起使用很多次，最后plt.show() 一起画出来 

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

#### plt.plot(,,color="")  第三个参数可以手动添加颜色，用于区分曲线

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')
plt.show()
```

#### 图像也可以设置样式 plt.plot(,,,linestyle="--"),但是这个--不能瞎写

```python
plt.plot(x,siny)
plt.plot(x,cosy,linestyle="-.")  #-. -- - :四种东西
plt.show()
```

#### 图像的范围也是可以调整的，自己的初始数据为 np.linspace(0,10,100) ,  值域是 (-1,1)
#### 使用 plt.xlim(a,b) 设定范围
#### 也可以使用一条语句 同时设定范围
#### plt.axis([-1,11,-2,2])  一个列表 4个数，前两个是x范围，后两个是 y范围

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="blue",linestyle=":")
plt.xlim(-5,15)   #设定x范围 
plt.ylim(0,1.5)    #设定y范围
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="blue",linestyle=":")
plt.axis([-1,11,-2,2])   
plt.show()
```

### 设定坐标标注，更好的看
### plt.xlabel("字符串")    plt.ylabel("字符串")

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="blue",linestyle=":")
plt.xlabel("x axis")    #设定X坐标标注
plt.ylabel("y axis")    #设定y坐标标注
plt.show()
```

#### plt.plot(x,y,label="xxx"),  但是输出前要写 plt.legend()

```python
plt.plot(x,siny,label="sin(x)")   #绘图时，加入label 字符串 
plt.plot(x,cosy,color="blue",linestyle=":",label="cos(x)" )
plt.xlabel("x axis")    
plt.ylabel("y axis")    
plt.legend()  #一定要写这个，写了标签之后
plt.show()
```

#### 对图像加入一个 标题  plt.title("xx")

```python
plt.plot(x,siny,label="sin(x)")   #绘图时，加入label 字符串 
plt.plot(x,cosy,color="blue",linestyle=":",label="cos(x)" )
plt.xlabel("x axis")    
plt.ylabel("y axis")    
plt.legend()  #一定要写这个，写了标签之后
plt.title("ts")  #图有一个标题
plt.show()
```

### 散点图也和重要
#### Scatter Plot   plt.scatter()
#### 散点图横纵轴是特征 ，折线图y轴是label ,x是特征

```python
plt.scatter(x,siny,label="sin(x)")
plt.scatter(x,cosy,label="cos(x)")
plt.xlabel("x axis")    
plt.ylabel("y axis")
plt.axis([-1,11,-2,2])
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

```python
x=np.random.normal(0,1,10000)   # 2维正态分布
y=np.random.normal(0,1,10000)   
plt.scatter(x,y,alpha=0.1)      #不透明度
plt.show()   #散点图
```
