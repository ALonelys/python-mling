```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris =datasets.load_iris()
iris      #封装的数据
```

```python
iris.keys()  #是一个字典
```

```python
print(iris.DESCR)    #文档，然后就是一点的是
```

```python
iris.data   #4列 特征，二维矩阵
```

```python
iris.data.shape 
```

```python
iris.feature_names  #特征量
```

```python
print(iris.target)  #结果答案
print(iris.target.shape)   #规模
print(iris.target_names)  #label 类型名称
```

```python
x=iris.data[:,:2]
print(x)
```

```python

plt.scatter(x[:,0],x[:,1],label="data")   #分别取出 0，1列
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("now ")
plt.legend()
plt.show()
```

```python
y=iris.target
print(x)
print(y)
```

#### plt.scatter(x,y)  其中装的是每一个点

```python
plt.scatter(x[y==0,0],x[y==0,1],color="red",marker="x")  #  这个用法有点不熟练
plt.scatter(x[y==1,0],x[y==1,1],color="blue",marker="o")
plt.scatter(x[y==2,0],x[y==2,1],color="green",marker="+")

plt.show()
```

#### 上面数据点有重合的


```python
x[y==0,0].size       #会有重复的点的，确实会有的
```

```python
x=iris.data[:,2:]
plt.scatter(x[y==0,0],x[y==0,1],color="red",marker="x") 
plt.scatter(x[y==1,0],x[y==1,1],color="blue",marker="o")
plt.scatter(x[y==2,0],x[y==2,1],color="green",marker="+")
plt.show()
```
